```python
import pandas as pd
import sqlite3

# Connect to SQLite database
conn = sqlite3.connect('database.db')

# Read CSV file into DataFrame
df = pd.read_csv('/Electric_Vehicle_Population_Data.csv')
df
```

```python
tosqltable= df.to_sql('Electric_Vehicle_Population_Data', conn, if_exists='replace', index=False)
tosqltable
```

```python
query = "SELECT * FROM Electric_Vehicle_Population_Data LIMIT 6"
result = pd.read_sql(query, conn)
print(result)
```

```python
# Execute the SQL query to get the count of vehicles by Make and Year

query1 = " SELECT make,model Year,count(*) FROM Electric_Vehicle_Population_Data GROUP BY 1,2 ORDER BY 3 DESC "
result1 = pd.read_sql(query1, conn)
print(result1)
```
